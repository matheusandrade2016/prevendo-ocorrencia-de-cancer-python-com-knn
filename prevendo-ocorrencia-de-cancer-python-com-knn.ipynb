```python
# Importando nossas Bibliotecas 

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# Carrehando nosso dataset
dados = pd.read_csv("/kaggle/input/ocorrncia-de-cncer/dataset.csv")
```

```python
# Verificando meus dados
dados.head()
```

```python
# Retirando minha coluna ID
# dados = dados.iloc[:,1:32]
dados = dados.drop(['id'], axis = 1)
```

```python
# Mudando os valores da coluna Diagnosis usando np.where
dados.diagnosis = np.where((dados.diagnosis == 'M'), 'Maligno', 'Benigno')
```

```python
# Verificando meus dados novamente
dados.head()
```

```python
# Contando meus valores da variavel
dados.diagnosis.value_counts()
```

```python
# Medidas de TendÃªncia Central

# Detectamos um problema de escala entre os dados, que entÃ£o p
# recisam ser normalizados
(dados[['radius_mean', 'area_mean', 'smoothness_mean']]).describe()
```

```python
# Criando nossas variaveis para a separaÃ§Ã£o dos dados
X = dados.drop(['diagnosis'], axis = 1)
y = dados.diagnosis
```

```python
# Criando nosso Escalonador para o teste
escalonador = StandardScaler()
```

```python
# Criando um novo Datraframe
dados_normalizados = pd.DataFrame(escalonador.fit_transform(X))
```

```python
# Dividindo nossos dados em treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.7, 
                                                    random_state = 0)
```

```python
# Criando meu Modelo KNN
modelo_knn = KNeighborsClassifier(n_neighbors= 3)
```

```python
# Fazendo meu Treinamento dos dados
modelo_knn.fit(X_train,y_train)
```

```python
# Fazendo minhas previsÃµes 
previsoes_knn = modelo_knn.predict(X_test)
print(previsoes_knn)
```

```python
# Verificando minha acuracia
acuracia_modelo = accuracy_score(y_test, previsoes_knn)
print()
print(f"Acuaraica do Modelo : {acuracia_modelo:.2f}%")
```
